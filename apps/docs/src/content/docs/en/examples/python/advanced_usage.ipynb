```python
"""Advanced Usage: AI-Generated Code Execution & Sandbox Workflows

This notebook demonstrates how AI-generated code can be:
- Generated dynamically
- Executed safely in a sandbox
- Validated automatically
- Improved via feedback loops

---

### Why this matters
Executing AI-generated code directly is risky. This notebook shows a controlled approach to make it safe and reliable."""
```

```python
import subprocess
import tempfile
import time
import os
```

```python
def generate_code(prompt: str) -> str:
    """
    Simulated AI code generation.
    Replace with LLM API in real use.
    """

    if "factorial" in prompt:
        return """
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))
"""

    elif "fibonacci" in prompt:
        return """
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(6))
"""

    elif "bug" in prompt:
        return "print(10/0)"  # intentional error

    return "print('Unsupported task')"
```

```python
def run_in_sandbox(code: str, timeout: int = 5):
    """
    Executes code safely using a temporary file.
    """

    try:
        with tempfile.NamedTemporaryFile(delete=False, suffix=".py") as tmp:
            tmp.write(code.encode())
            file_path = tmp.name

        start = time.time()

        result = subprocess.run(
            ["python", file_path],
            capture_output=True,
            text=True,
            timeout=timeout
        )

        end = time.time()

        return {
            "output": result.stdout.strip(),
            "error": result.stderr.strip(),
            "time": round(end - start, 4)
        }

    except subprocess.TimeoutExpired:
        return {"output": "", "error": "Execution timed out", "time": timeout}

    except Exception as e:
        return {"output": "", "error": str(e), "time": 0}
```

```python
def validate(prompt: str, output: str) -> bool:
    """
    Simple correctness check.
    """

    expected = {
        "factorial": "120",
        "fibonacci": "8"
    }

    for key in expected:
        if key in prompt:
            return output == expected[key]

    return False
```

```python
def execute_pipeline(prompt: str):
    print(f"\n🔹 Prompt: {prompt}")

    # Step 1: Generate
    code = generate_code(prompt)
    print("\nGenerated Code:\n", code)

    # Step 2: Execute
    result = run_in_sandbox(code)

    print("\nOutput:", result["output"])
    print("Error:", result["error"])
    print("Execution Time:", result["time"], "sec")

    # Step 3: Validate
    if result["output"]:
        if validate(prompt, result["output"]):
            print("✅ Validation Passed")
        else:
            print("❌ Validation Failed")

    return result
```

```python
# Successful case
execute_pipeline("factorial")

# Another valid case
execute_pipeline("fibonacci")

# Error case
execute_pipeline("bug")
```

```python
def retry_execution(prompt: str, retries: int = 2):
    for i in range(retries):
        print(f"\nAttempt {i+1}")

        result = execute_pipeline(prompt)

        if not result["error"]:
            print("✅ Success achieved")
            return result

        print("⚠️ Retrying with fallback...")

    print("❌ Failed after retries")
```

```python
# Test retry logic
retry_execution("bug")
```
